```python
import fastf1
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

print("All libraries loaded successfully!")
print(f"FastF1 version: {fastf1.__version__}")
```

```python
# Set up cache folder
fastf1.Cache.enable_cache('f1_cache')

# Load the most recent 2025 race to test
session = fastf1.get_session(2025, 1, 'R')
session.load()

print("Data loaded!")
print(f"Race: {session.event['EventName']}")
print(f"Winner: {session.results.iloc[0]['FullName']}")
```

```python
# Pull all 2025 races completed so far
races_data = []

# 2025 races so far (first 4 rounds)
rounds = [1, 2, 3, 4]

for round_num in rounds:
    try:
        session = fastf1.get_session(2025, round_num, 'R')
        session.load(telemetry=False, weather=False, messages=False)
        
        results = session.results[['FullName', 'TeamName', 'GridPosition', 'Position', 'Points']].copy()
        results['Round'] = round_num
        results['RaceName'] = session.event['EventName']
        races_data.append(results)
        print(f"✓ Round {round_num}: {session.event['EventName']}")
    except Exception as e:
        print(f"✗ Round {round_num}: {e}")

# Combine all races
df = pd.concat(races_data, ignore_index=True)
print(f"\nTotal rows: {len(df)}")
print(df.head())
```

```python
all_data = []

print("Loading 2025 season...")
for round_num in range(1, 25):
    try:
        session = fastf1.get_session(2025, round_num, 'R')
        session.load(telemetry=False, weather=False, messages=False)
        
        results = session.results[['FullName', 'TeamName', 'GridPosition', 'Position', 'Points']].copy()
        results['Round'] = round_num
        results['Year'] = 2025
        results['RaceName'] = session.event['EventName']
        all_data.append(results)
        print(f"  ✓ Round {round_num}: {session.event['EventName']}")
    except Exception as e:
        print(f"  ✗ Round {round_num}: skipped ({e})")

df_all = pd.concat(all_data, ignore_index=True)
print(f"\nTotal rows: {len(df_all)}")
```

```python
data_2026 = []

print("Loading 2026 season...")
for round_num in range(1, 4):
    try:
        session = fastf1.get_session(2026, round_num, 'R')
        session.load(telemetry=False, weather=False, messages=False)
        
        results = session.results[['FullName', 'TeamName', 'GridPosition', 'Position', 'Points']].copy()
        results['Round'] = round_num
        results['Year'] = 2026
        results['RaceName'] = session.event['EventName']
        data_2026.append(results)
        print(f"  ✓ Round {round_num}: {session.event['EventName']}")
    except Exception as e:
        print(f"  ✗ Round {round_num}: skipped ({e})")

df_2026 = pd.concat(data_2026, ignore_index=True)
print(f"\n2026 rows: {len(df_2026)}")
```

```python
# Combine 2025 and 2026
df_combined = pd.concat([df_all, df_2026], ignore_index=True)

# Save to SQL
engine = create_engine('sqlite:///f1_2025.db')
df_combined.to_sql('race_results', engine, if_exists='replace', index=False)

print(f"Total rows: {len(df_combined)}")
print(f"Breakdown:\n{df_combined.groupby('Year')['Round'].nunique()}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error

le_driver = LabelEncoder()
le_team = LabelEncoder()

df_model = df_combined.dropna(subset=['GridPosition', 'Position']).copy()
df_model['Driver_enc'] = le_driver.fit_transform(df_model['FullName'])
df_model['Team_enc'] = le_team.fit_transform(df_model['TeamName'])

# Give 2026 data 5x more weight than 2025
df_model['weight'] = df_model['Year'].apply(lambda x: 3.0 if x == 2026 else 1.0)

X = df_model[['GridPosition', 'Driver_enc', 'Team_enc', 'Round', 'Year']]
y = df_model['Position']
w = df_model['weight']

X_train, X_test, y_train, y_test, w_train, w_test = train_test_split(
    X, y, w, test_size=0.2, random_state=42
)

# Train with sample weights
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train, sample_weight=w_train)

predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
print(f"Model trained with 2026 weighting!")
print(f"Mean Absolute Error: {mae:.2f} positions")
```
